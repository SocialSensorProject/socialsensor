```python
df = pd.read_csv('data_violinplots/Cele_death_Dataset_hashtag_features.txt', sep='\t')
df = df.drop(columns=['id','hashtag1','hashtag2','hashtag3','userCount']).sort_values("tweetCount")
#df.drop(df.index[:200000], inplace=True)
#df = df.head(50000)
scatter = df

df1 = df[df['tweetCount'].between(65, 100, inclusive=True)]
df1 = df1.drop(columns=['tweetCount'])

df2 = df[df['tweetCount'].between(101, 1000, inclusive=True)]
df2 = df2.drop(columns=['tweetCount'])

df3 = df[df['tweetCount'].between(1001, 10000, inclusive=True)]
df3 = df3.drop(columns=['tweetCount'])

df4 = df[df['tweetCount'].between(10001, 100000, inclusive=True)]
df4 = df4.drop(columns=['tweetCount'])

df5 = df[df['tweetCount'].between(100001, 1000000, inclusive=True)]
df5 = df5.drop(columns=['tweetCount'])

```
